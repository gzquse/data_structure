```python
import requests
from bs4 import BeautifulSoup
```

```python
base_url = 'https://cd.lianjia.com'
url = 'https://cd.lianjia.com/zufang/shuangliu/rp1/'

def get_links(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.110 Safari/537.36'}
    response = requests.get(url,headers = headers)
    soup = BeautifulSoup(response.text, 'lxml')
    links = soup.find_all('a',target = '_blank')
    link_lists = [link.get('href') for link in links[12:]]
    link_lists = [base_url + link for link in link_lists]
    return link_lists
```
