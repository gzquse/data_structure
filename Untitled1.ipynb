```python
str1 = "1234567"
def trans(str1):
    l = list(str1)
    for i in range(len(str1), 0, -3):
        if i != len(str1):
            l.insert(i, ',')
    ret = ''.join(l)
    return ret
trans(str1)
```

```python
class Solution:
    def removeDuplicateLetters(self, s: str) -> str:
        seen = set()
        stack = []
        remain_c = collections.Counter(s)
        for c in s:
            if c not in seen:
                while stack and c < stack[-1] and remain_c[stack[-1]] > 0:
                    seen.discard(stack.pop())
                seen.add(c)               
                stack.append(c)
                print(seen)
                print(stack)
            remain_c[c] -= 1
        return ''.join(stack)
    
"bcabc"
{'b'}
['b']
{'c', 'b'}
['b', 'c']
{'a'}
['a']
{'b', 'a'}
['a', 'b']
{'c', 'b', 'a'}
['a', 'b', 'c']
```
